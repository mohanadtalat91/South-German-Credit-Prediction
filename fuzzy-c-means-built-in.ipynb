```python
import pandas as pd
from pyclustering.cluster.fcm import fcm
import numpy as np
from matplotlib import pyplot as plt
from pyclustering.cluster.center_initializer import kmeans_plusplus_initializer
```

```python
train = pd.read_csv('train.csv')
```

```python
print(train.head())
print(train.info())
print(train.describe())
```

```python
X= train.drop(['Id','kredit', 'gastarb', 'telef', 'laufkont', 'pers'], axis=1).to_numpy()
X
```

```python
initial_centers = kmeans_plusplus_initializer(X, 2, kmeans_plusplus_initializer.FARTHEST_CENTER_CANDIDATE).initialize()
```

```python
FCM = fcm(X,initial_centers)
```

```python
FCM.process()
member = FCM.get_membership()
```

```python
print(member)
```

```python
def alpha_cut(thereshold,member):
  x= list()
  for i,j in member:
    if i > 0.50: x.append(1)
    else : x.append(0)
  return x
```

```python
data_alpha = alpha_cut(0.5,member)
print(data_alpha)
```

```python
def acc(x_pred, y):
  acc=0
  for i in range(len(x_pred)):
    if x_pred[i] == y[i]:
      acc= acc+1
  acc= acc/len(x_pred)
  return acc
```

```python
y= train[['kredit']].to_numpy()
y
```

```python
ac= acc(data_alpha,y)
```
